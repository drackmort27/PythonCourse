```python
def doblar_valor(numero):
    numero *= 2
n = 10
doblar_valor(n)
n
```

```python
def doblar_valores(numeros):
    for i,n in enumerate(numeros):
        numeros[i] *= 2
ns = [10,50,100]
doblar_valores(ns)
ns    #para numeros normales los cambios que se generan en la funcion solo quedan en la funcion (por valor), encambio como por ejemplo en las listas
      #es por referencia y cambia el valor de la variable original
```

```python
def doblar_valor(numero):
    return numero*2
n = 10
doblar_valor(n)
n    #Asi si podemos modificarlo
```

```python
#Para evitar en la parte de referencia se crea una copia en la llamada
def doblar_valores(numeros):
    for i,n in enumerate(numeros):
        numeros[i] *= 2
ns = [10,50,100]
doblar_valores(ns[:]) #Aqui
ns  
```
