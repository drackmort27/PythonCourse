```python
def cuenta_atras(num):
    num -= 1
    if num > 0:
        print(num)
        cuenta_atras(num) #Lo mas importante de una funcion recursiva es que se deje de llamar en algun momento
    else:
        print("Boom")
    print(f"Fin de la funcion {num}")
cuenta_atras(5)
```

```python
def sumatorio(numero):
    if numero > 0:
        # Sumamos el resultado devuelto por la propia llamada pasando el número menos 1
        numero += sumatorio(numero-1)
    # Finalmente devolvemos el numero
    return numero
sumatorio(3)
```
