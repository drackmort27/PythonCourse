```python
def suma(a,b): #Durante la definicion a y b se denominan parametros 
    return a+b
suma(2,4) #Durante la llamada a y b se denominan argumentos
```

```python
def par_o_impar(numero):
    if numero % 2 == 0:
        print("PAR")
    else:
        print("IMPAR")
par_o_impar(3)
```

```python
def recortar(numero,minimo,maximo):
    if numero < minimo:
        return minimo
    elif numero > maximo:
        return maximo
    else:
        return numero
recortar(6,3,5)
```
