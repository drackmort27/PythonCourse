```python
sumatorio = 0
numero = 7
# Completa el ejercicio aquí

for i in range(numero):
    if i % 5 != 0 and i % 7 != 0:
        sumatorio += i
        print(sumatorio)

print(f"La suma da: {sumatorio}")
```

```python
iniciales = ["Hola", "Mundo"]

for i,c in enumerate(iniciales):
    iniciales[i] = iniciales[i][0]
    
    print(iniciales)
    print(i,c)
    print(c[0])

print(iniciales)
```

```python
multiplos = []
# Completa el ejercicio 
numero = int(input("Ingrese un numero entero del 1 al 9: "))

while numero > 9:
    numero = int(input("Error,ingrese un numero entero del 1 al 9: "))

for i in range(100):
    if i % numero == 0:
        multiplos.append(i)
print(f"Los multiplos de {numero} son {multiplos}")
```
