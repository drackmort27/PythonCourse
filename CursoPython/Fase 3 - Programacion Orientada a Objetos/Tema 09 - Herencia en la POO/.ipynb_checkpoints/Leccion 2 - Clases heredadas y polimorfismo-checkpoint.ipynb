```python
class Producto:
    def __init__(self,referencia,nombre,pvp,descripcion):
        self.referencia = referencia
        self.nombre = nombre
        self.pvp = pvp
        self.descripcion = descripcion
        
    def __str__(self):
        return """\
REFERENCIA\t{}
NOMBRE\t\t{}
PVP\t\t{}
DESCRIPCIÓN\t{}""".format(self.referencia,self.nombre,self.pvp,self.descripcion)
    

class Adorno(Producto):
    pass


class Alimento(Producto):
    productor = ""
    distribuidor = ""
    
    def __str__(self):
        return """\
REFERENCIA\t{}
NOMBRE\t\t{}
PVP\t\t{}
DESCRIPCIÓN\t{}
PRODUCTOR\t{}
DISTRIBUIDOR\t{}""".format(self.referencia,self.nombre,self.pvp,self.descripcion,self.productor,self.distribuidor)


class Libro(Producto):
    isbn = ""
    autor = ""
    
    def __str__(self):
        return """\
REFERENCIA\t{}
NOMBRE\t\t{}
PVP\t\t{}
DESCRIPCIÓN\t{}
ISBN\t\t{}
AUTOR\t\t{}""".format(self.referencia,self.nombre,self.pvp,self.descripcion,self.isbn,self.autor)
```

```python
a = Adorno(2034,"Vaso",15,"Vaso con arboles")

al = Alimento(2035,"Botella de aceite",5,"250ml")
al.productor= "La aceitera"
al.distribuidor = "Distribuciones SA"

li = Libro(2037,"Harry Potter",6,"El misterio del principe")
li.isbn= "0-123123134-7"
li.autor = "J.K.ROWLING"
```

```python
productos = [a,al]
```

```python
productos.append(li)
```

```python
productos
```

```python
for p in productos:
    print(p,"\n")
```

```python
for p in productos:
    print(p.referencia,p.nombre)
```

```python
for p in productos:
    if (isinstance(p, Adorno) ):
        print(p.referencia,p.nombre)
    elif(isinstance(p, Alimento)):
        print(p.referencia,p.nombre,p.productor)
    elif(isinstance(p, Libro)):
        print(p.referencia,p.nombre,p.autor)
```

```python
def rebajar_producto(p, rebaja):
    """Devuelve un producto en rebaja en porcentaje con su precio"""
    p.pvp = p.pvp - (p.pvp/100 * rebaja)
    return p

a_rebajado = rebajar_producto(a, 10)
print(a_rebajado)
```

```python
#Para copiar un objeto y no cambiarle el valor
import copy
copia_a = copy.copy(a)
```

```python
print(copia_a)
```

```python
copia_a.pvp = 25
```

```python
print(copia_a)
print(a)
```

```python
#Polimorfia:Propiedad de la herencia en que los objetos de distintas subclases pueden responder una misma accion
#El metodo rebajar_producto() recibe objetos de distintas clases y accede al atributo pvp dando por hecho que existe en ellos
#En python todas las clases son a su vez subclases de la superclase Objetc, es decir, son polimorficas por defecto
```
