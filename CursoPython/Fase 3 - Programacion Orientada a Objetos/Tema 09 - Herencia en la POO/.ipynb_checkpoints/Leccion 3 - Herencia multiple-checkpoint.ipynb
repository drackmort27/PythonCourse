```python
#Herencia multiple: una subclase puede heredar de multiples superclases a la vez (esto cuando tienen metodos y atributos en comun)
#Python en este caso da prioridad a la superclase mas a la izquierda
```

```python
class A:
    def __init__(self):
        print("clase A")
    def a(self):
        print("heredo este metodo de a")
class B:
    def __init__(self):
        print("clase B")
    def b(self):
        print("heredo este metodo de b")
class C(B,A):
    def c(self):
        print("metodo de c")
c = C()
```

```python
c.a()
c.b()
c.c()
```
