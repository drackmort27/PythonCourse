```python
class Galleta():
    chocolate = False #Atributo de clase
    def __init__(self):  #El metodo init es una funcion interna de la clase y un metodo es la palabra que indica una funcion dentro de la clase
        print("Se acaba de crear una galleta.")   #Y se comparte con todos los objetos de la clase y permite enviarle argumentos en la instanciacion
                         #Self hace referencia al propio objeto, sirve para diferenciar entre el ambito de clase y el de un metodo
                         #Es implicito en los metodos en la definicion
    def chocolatear(self):
        self.chocolate = True #Esto para poder referenciarnos a un atributo interno de la clase, si no no funciona
    def tiene_chocolate(self):
        if (self.chocolate):
            print("Soy una galleta choc")
        else:
            print("Soy una galleta sin choc")
#Gracias al self puedo hacer referencia al propio objeto
g = Galleta()
g.tiene_chocolate()
g.chocolatear()
g.chocolate
g.tiene_chocolate()
```
