```python
#La encapsulacion es una funcionalidad para impedir el acceso externo a atributos y metodos
#Python no la tiene pero se puede simular:

class Ejemplo:
    __atributo_privado = "Soy un atributo inalcanzable desde fuera"

    def __metodo_privado(self):
        print("Soy un metodo inalcanzable desde fuera")
```

```python
e = Ejemplo()
```

```python
e.__atributo_privado
```

```python
e.__metodo_privado()
```

```python
class Ejemplo:
    __atributo_privado = "Soy un atributo inalcanzable desde fuera"

    def __metodo_privado(self):
        print("Soy un metodo inalcanzable desde fuera")

    def atributo_publico(self):
        return self.__atributo_privado
    def metodo_publico(self):
        return self.__metodo_privado()
```

```python
e = Ejemplo()
```

```python
e.atributo_publico()
```

```python
e.metodo_publico()
```
