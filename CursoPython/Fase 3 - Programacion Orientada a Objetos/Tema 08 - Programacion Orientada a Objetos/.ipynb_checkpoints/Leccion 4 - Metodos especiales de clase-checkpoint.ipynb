```python
class Pelicula:
    #Constructor de clase
    def __init__(self,titulo,duracion,lanzamiento):
        self.titulo = titulo
        self.duracion = duracion
        self.lanzamiento = lanzamiento
        print(f"Se creo la pelicula {self.titulo}")

    #Destructor de clase
    def __del__(self):
        print("Se esta borrando la pelicula",self.titulo)

    #Redefinimos el metodo string
    def __str__(self):
        return f"{self.titulo} lanzada en {self.lanzamiento} con una duracion de {self.duracion} minutos"
p = Pelicula("Star Wars",180,1972)
```

```python
class Pelicula:
    #Constructor de clase
    def __init__(self,titulo,duracion,lanzamiento):
        self.titulo = titulo
        self.duracion = duracion
        self.lanzamiento = lanzamiento
        print(f"Se creo la pelicula {self.titulo}")

    #Destructor de clase
    def __del__(self):
        print("Se esta borrando la pelicula",self.titulo)

    #Redefinimos el metodo string
    def __str__(self):
        return f"{self.titulo} lanzada en {self.lanzamiento} con una duracion de {self.duracion} minutos"

    #Redefinimos len
    def __len__(self):
        return self.duracion
        
p = Pelicula("Star Wars",180,1972)
```
