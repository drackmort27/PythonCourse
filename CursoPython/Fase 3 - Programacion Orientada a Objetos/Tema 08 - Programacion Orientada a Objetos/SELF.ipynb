```python
class A:
    def __init__(argumento):   #Es la propia instancia del objeto en la memoria el self
        argumento.atributo = "Hola que tal"

    def algo(cosa):
        cosa.atributo = "Adios que tal"
```

```python
a = A()
```

```python
a.atributo
```

```python
a.algo()
```
